```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Создаем сигнал во временной области
sample_rate = 1000  # Частота дискретизации в Гц
t = np.linspace(0, 1, sample_rate, endpoint=False)  # Временной вектор от 0 до 1 секунды
frequency = 5  # Частота сигнала в Гц
signal = np.sin(2 * np.pi * frequency * t)
```

```python
len(signal),signal
```

```python
# Выполняем преобразование Фурье для получения частотного представления
signal_freq = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(signal), 1 / sample_rate)
```

```python
# Изменяем фазу сигнала, например, на 90 градусов
phase_shift = np.pi / 2
signal_freq_shifted = np.abs(signal_freq) * np.exp(1j * (np.angle(signal_freq) + phase_shift))

# Инвертирование обратно во временную область
signal_shifted = np.fft.ifft(signal_freq_shifted).real

# Визуализация исходного и смещенного сигналов
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Исходный сигнал')

plt.subplot(2, 1, 2)
plt.plot(t, signal_shifted)
plt.title('Сигнал со смещенной фазой (на 90 градусов)')

# plt.subplot(2, 2, 2)
# plt.plot(t, signal_shifted)
# plt.title('Сигнал со смещенной фазой (на 90 градусов)')


plt.tight_layout()
plt.show()
```
